1. Завантажте набір даних books.csv.  

```python
import pandas as pd

books = pd.read_csv('books.csv')
books.head(4)
```

2. Створіть стовпці з кількістю голосних в імені автора.

```python
import re

books['Vowel Count'] = books['Author'].str.split(', ').str.get(-1).str.count('(?i)[aeiouy]')
books.head(4)
```

3. Знайдіть авторів із прізвищами з 3 голосними поспіль.

```python
books[books['Author'].str.split(',').str.get(0).str.contains('[aeiouy]{3}', na=False)]
```

4. Створіть стовпець і внесіть туди 1, якщо в назві є цифри, і 0 у протилежному випадку.  

```python
books['Has number'] = books['Title'].str.contains(r'[0-9]').apply(lambda x: 1 if x else 0)
books.iloc()[166:170]
```

5. Знайдіть найдовший заголовок у наборі даних. 

```python
books[books['Title'].str.len() == books['Title'].str.len().max()]
```

6. Знайдіть книги, написані автором, ім’я якого починається на „A”  

```python
books[books['Author'].str.split(', ').str.get(-1).str.startswith('A', na=False)].head(4)
```

7. Знайдіть книги, видані видавництвами, у назві яких 2 слова. 

```python
books[books['Publisher'].str.split().str.len() == 2].head(4)
```

8. Знайдіть авторів із 3-м «у» у прізвищі
